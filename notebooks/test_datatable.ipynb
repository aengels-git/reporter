```python
tab = prep_freq_table(diamonds, ["cut","color"])
digit = 2
percentage = True
if percentage:
    format_mapping = {col: f"{{:,.{digit}%}}" for col in tab.filter(regex="^prop",axis="columns").columns}
else:
    format_mapping = {col: f"{{:,.{digit}f}}" for col in tab.filter(regex="^prop",axis="columns").columns}
for key, value in format_mapping.items():
    tab[key] = tab[key].apply(value.format)
tab
```

```python
def highlight_cols(x):
    #copy df to new - original data are not changed
    df = x.copy()
    #select all values to default value - red color
    df.loc[:,:] = 'background-color: red'
    #overwrite values grey color
    df[['cut','counts']] = 'background-color: grey'
    #return color df
    return df  
prep_freq_table(diamonds, ["cut","color"]).apply(highlight_cols, axis=None)
```
